# **Drew the curves of O(n) notation**

---




```
import matplotlib.pyplot as plt
import numpy as np
import time


def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1


def plot_merge_sort(n):
    fig, ax = plt.subplots()

    # Generate a list of colors based on the length of n
    colors = plt.cm.rainbow(np.linspace(0, 5, 8))
     
    for i, size in enumerate(n):
        arr = np.random.randint(1000, size=size)
        start = time.time()
        merge_sort(arr)
        end = time.time()
        ax.plot([size], [end - start], color=colors[i], label=f'n = {size}')  # Plot a single point for each array size
        ax.plot([0, size], [0, end - start], color=colors[i])  # Plot a line connecting time to n

    ax.set(xlabel='n', ylabel='time', title='Merge Sort')
    ax.legend()
    plt.show()

plot_merge_sort([550000, 750000, 950000]) # the length of curve
```

# **Measure the execution time in milliseconds**

```

import random
import time

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

def test_merge_sort():
    best_case = list(range(1000)) # already sorted list
    average_case = [random.randint(1, 5000) for _ in range(5000)] # random list of integers
    worst_case = [random.randint(1, 10000) for _ in range(10000)]
    
    for case in [best_case,average_case, worst_case]:
        start_time = time.time()
        merge_sort(case)
        end_time = time.time()
        print(f"Sorting {len(case)} elements in {end_time - start_time:.6f} milliseconds")
```

```
test_merge_sort()
```
